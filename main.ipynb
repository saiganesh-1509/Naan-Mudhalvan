```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the preprocessed data
data = pd.read_csv("preprocessed_data.csv")

# List of columns to visualize
columns_to_visualize = [
    'Worked for 3 months or more but less than 6 months -  Persons',
    'Worked for 3 months or more but less than 6 months - Males',
    'Worked for 3 months or more but less than 6 months - Females',
    'Worked for less than 3 months - Persons',
    'Worked for less than 3 months - Males',
    'Worked for less than 3 months - Females',
    'Industrial Category - A - Cultivators - Persons',
    'Industrial Category - A - Cultivators - Males',
    'Industrial Category - A - Cultivators - Females',
    # Add the rest of the columns you want to visualize here
]

# Create a bar plot for each column
for column in columns_to_visualize:
    plt.figure(figsize=(10, 6))
    plt.bar(data['Age group'], data[column])
    plt.title(f'{column} by Age Group')
    plt.xlabel('Age Group')
    plt.ylabel(column)
    plt.xticks(rotation=45)
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a histogram of the original column
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x=data['Worked for 3 months or more but less than 6 months -  Persons'], kde=True)
plt.title('Distribution of Worked for 3-6 Months - Persons')
plt.xlabel('Count')
plt.ylabel('Density')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, y=data['Worked for 3 months or more but less than 6 months - Males'])
plt.title('Box Plot of Worked for 3-6 Months - Persons')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x=data['Worked for 3 months or more but less than 6 months -  Persons'], y=data['Worked for 3 months or more but less than 6 months - Males'])
plt.title('Scatter Plot of Worked for 3-6 Months - Persons vs. Worked for 3-6 Months - Males')
plt.xlabel('Worked for 3-6 Months - Persons')
plt.ylabel('Worked for 3-6 Months - Males')
plt.show()
```

```python
columns_to_correlate = [
    'Worked for 3 months or more but less than 6 months -  Persons',
    'Worked for less than 3 months - Persons',
    'Industrial Category - A - Cultivators - Persons'
]
correlation_matrix = data[columns_to_correlate].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
